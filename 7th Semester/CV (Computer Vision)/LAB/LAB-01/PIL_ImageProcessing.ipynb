#### Importing Libraries

```python
from PIL import Image
from IPython.display import display
```

#### Opening and displaying Image

```python
im = Image.open("cat.jpg")
display(im)
```

```python
gray_img = Image.open('cat.jpg').convert('L')
display(gray_img)
```

#### Rotate an image:

The following script rotates an image 45 degrees, and displays it

```python
im_rotated = im.rotate(45)
display(im_rotated)
```

#### Saving Image after changes

```python
im_rotated.save('rotated_cat.png')
```

#### Attributes of Image Module

The instance of the Image class has some attributes.

```python
print("Image Filename: ", im.filename)
print("Image format: ", im.format)
print("Image mode: ", im.mode)
print("Image width: ", im.width)
print("Image height: ", im.height)
print("Image size: ", im.size)
```

#### Resize an image

```python
#Make the new image half the width and half the height of the original image
resized_image = im.resize((im.size[0] // 2, im.size[1] // 2))

#Display the resized imaged
display(resized_image)

#Save the resized image
resized_image.save('resized_image.png')
```

#### Creating Thumbnails

Sometimes, it is required to have all images of equal height and width. One way to achieve this, is by creating a thumbnail of all images using the thumbnail() function from PIL.

```python
im_rotated.thumbnail((80,80))
im_rotated.save('cat_thumbnail.png')
thumbnail_im = Image.open('cat_thumbnail.png')
display(thumbnail_im)
```

#### Splitting Image bands

Image.split() method is used to split the image into individual bands. This method returns a tuple of individual image bands from an image.

```python
r, g, b = im.split()
```

#### Merging Images

The merge() function accepts a mode and a tuple of images as parameters, and combines them into a single image.

```python
image = Image.merge("RGB", (b, g, r))
display(image)
```

```python
image2 = Image.merge("RGB", (g, r, b))
display(image2)
```

#### Image Colors

The ImageColor module supports the following strings formats:

+ Hexadecimal color specifiers, given as rgb
+ HSL (Hue, Saturation, Lightness)

##### getrgb()


Convert a color string to an RGB tuple. If the string cannot be parsed, this function raises a ValueError exception. 

```python
from PIL import ImageColor


# using getrgb
red_rgb = ImageColor.getrgb("red")
print(red_rgb)

blue_rgb = ImageColor.getrgb("blue")
print(blue_rgb)

purple_rgb = ImageColor.getrgb("purple")
print(purple_rgb)

orange_rgb = ImageColor.getrgb("orange")
print(orange_rgb)
```

```python
# Create new image & get color RGB tuple.
img = Image.new("RGB", (128, 128), ImageColor.getrgb("#00ff00"))

#Show image
display(img)
```

#### Blur an Image

Blurring an image can be done by applying a filter to an image. The ImageFilter class in the Pillow library provides several standard image filters. Image filters can be applied to an image by calling the filter() method of Image object with required filter type as defined in the ImageFilter class.

```python
# Original Image
display(im)
```

```python
from PIL import ImageFilter

blur_image = im.filter(ImageFilter.BLUR)
display(blur_image)

#Save blur_image
blur_image.save('blur_image.png')
```

#### Crop an Image

The crop() function of the image class in Pillow requires the portion to be cropped as rectangle. The rectangle portion to be cropped from an image is specified as a four-element tuple and returns the rectangle portion of the image that has been cropped as an image Object.

```python
# Setting the points for cropped image

left = 100
top = 65
right = 250
bottom = 200
cropped_image = im.crop((left, top, right, bottom))

#Display the cropped_image
display(cropped_image)

#Save the cropped image
cropped_image.save('cropped_image.png')
```

#### Flip an Image

Image module of the pillow library allows us to flip an image. Flipping an image is also useful for data augmentation. We are going to use the transpose function from the Image module for flipping the images.

##### Horizontal Flip

```python
h_flipped_image = im.transpose(Image.FLIP_LEFT_RIGHT)

# Show the horizontal flipped image
display(h_flipped_image)
```

##### Vertical Flip

```python
v_flipped_image = im.transpose(Image.FLIP_TOP_BOTTOM)

# Show the horizontal flipped image
display(v_flipped_image)
```

#### Drawing on Image
The ‘ImageDraw’ module provides simple 2D graphics support for Image Object.

#### Draw Text on Image

```python
from PIL import ImageFont

img_to_draw = Image.open("cat.jpg")

draw = ImageDraw.Draw(img_to_draw)

# specified font style and size
font = ImageFont.truetype('font.ttf', 20)
text = 'Computer Vision'

# drawing text size
draw.text((5, 5), text, font = font, align ="center")

display(img_to_draw)
```

#### Draw line on Image

```python
from PIL import ImageDraw

img_to_draw = Image.open("cat.jpg")

#Draw line
draw = ImageDraw.Draw(img_to_draw)
draw.line((0, 0) + img_to_draw.size, fill=(0,0,255))

#Show image
display(img_to_draw)
```

#### Draw Shapes using PIL 

#### Draw line

```python
line_img = Image.new('RGB', (500, 300), (125, 125, 125))
draw = ImageDraw.Draw(line_img)
draw.line((200, 100, 300, 200), fill=(0, 0, 0), width=10)

display(line_img)
```

#### Draw Ellipse

```python
ellipse_img = Image.new('RGB', (500, 300), (125, 125, 125))
draw = ImageDraw.Draw(ellipse_img)

draw.ellipse((200, 125, 300, 200), fill=(0, 0, 255))
display(ellipse_img)
```

#### Draw Rectangle

```python
rectangle_img = Image.new('RGB', (500, 300), (125, 125, 125))
draw = ImageDraw.Draw(rectangle_img)

draw.rectangle((200, 125, 300, 200), fill=(255, 0, 0))
display(rectangle_img)
```

#### Draw Polygon

```python
polygon_img = Image.new('RGB', (500, 300), (125, 125, 125))
draw = ImageDraw.Draw(polygon_img)

draw.polygon(((200, 200), (300, 100), (250, 50)), fill=(0, 255, 0))

display(polygon_img)
```
