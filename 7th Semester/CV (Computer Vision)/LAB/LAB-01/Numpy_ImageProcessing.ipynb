#### Importing required libraries

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
```

```python
#   to read the images
im = plt.imread("cat.jpg")

#to display the images
plt.imshow(im)
```

#### Image Details

```python
print('# of dims: ',im.ndim)
print("Type of image: ", type(im))
print("Shape of Image: ", im.shape)
print('Dtype: ',im.dtype)
```

#### Rotate Image by 90 degrees

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(im)
plt.title("Original")

fig.add_subplot(1, 2, 2)
plt.imshow(np.rot90(im))
plt.title("Rotated")
```

#### Filp Image

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
#   Up down flip
plt.imshow(np.flipud(im))
plt.title("Up down flip")

fig.add_subplot(1, 2, 2)
#   Left Right Flip
plt.imshow(np.fliplr(im))
plt.title("Left Right flip")
```

#### Visualize RGB channels

```python
img_R, img_G, img_B = im.copy(), im.copy(), im.copy()

img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0

img_rgb = np.concatenate((img_R,img_G,img_B), axis=1)

plt.figure(figsize=(15, 15))
plt.imshow(img_rgb)
```

#### Trim Image

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(im)
plt.title('Original')

img0 = im[100:200, 0:150, :]
fig.add_subplot(1, 2, 2)
plt.imshow(img0)
plt.title('Trimmed')
```
