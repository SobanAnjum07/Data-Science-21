### Questin 01
Scientific Data Analysis with NumPy You are a scientist conducting research on climate change. You have collected temperature data from weather stations around the world stored in a NumPy array. Your task is to analyze the data using NumPy to:
1. Calculate the average temperature for each month across all stations.
2. Identify the station with the highest and lowest average temperature.
3. Determine the month with the largest temperature variation.
4. Visualize the temperature data using line plots or heatmaps.



<b>Use the following data set

```python
import numpy as np

# Sample temperature data for each month (in Celsius)
temperature_data = np.array([
    # January
    [10, 12, 15, 8, 9],
    # February
    [11, 13, 14, 9, 10],
    # March
    [12, 14, 16, 10, 11],
    # April
    [14, 16, 18, 12, 13],
    # May
    [16, 18, 20, 14, 15],
    # June
    [18, 20, 22, 16, 17],
    # July
    [20, 22, 24, 18, 19],
    # August
    [22, 24, 26, 20, 21],
    # September
    [20, 22, 24, 18, 19],
    # October
    [18, 20, 22, 16, 17],
    # November
    [15, 17, 19, 13, 14],
    # December
    [12, 14, 16, 10, 11]
])

# Station names for reference
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']
```

```python
# Calculate the average temperature for each month across all stations
monthly_avg_temps = np.mean(temperature_data, axis=1)
print("Average temperature for each month across all stations:", monthly_avg_temps)
```

```python
# Calculate the average temperature for each station across all months
station_avg_temps = np.mean(temperature_data, axis=0)

# Identify the station with the highest and lowest average temperature
max_station_index = np.argmax(station_avg_temps)
min_station_index = np.argmin(station_avg_temps)

station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']

print(f"Station with the highest average temperature: {station_names[max_station_index]} ({station_avg_temps[max_station_index]:.2f} °C)")
print(f"Station with the lowest average temperature: {station_names[min_station_index]} ({station_avg_temps[min_station_index]:.2f} °C)")
```

```python
# Calculate the temperature variation for each month (max - min)
monthly_temp_variation = np.ptp(temperature_data, axis=1)

# Identify the month with the largest temperature variation
max_variation_month_index = np.argmax(monthly_temp_variation)

month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

print(f"Month with the largest temperature variation: {month_names[max_variation_month_index]} ({monthly_temp_variation[max_variation_month_index]} °C)")
```

```python
import matplotlib.pyplot as plt

# Line plot for average temperatures of each month
plt.figure(figsize=(10, 6))
plt.plot(month_names, monthly_avg_temps, marker='o')
plt.title('Average Monthly Temperature Across All Stations')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.show()


```

```python
# Heatmap for temperature data
plt.figure(figsize=(10, 6))
plt.imshow(temperature_data, cmap='hot', interpolation='nearest', aspect='auto')
plt.colorbar(label='Temperature (°C)')
plt.xticks(ticks=np.arange(len(station_names)), labels=station_names)
plt.yticks(ticks=np.arange(len(month_names)), labels=month_names)
plt.title('Temperature Data Heatmap')
plt.xlabel('Station')
plt.ylabel('Month')
plt.show()
```

```python
# Calculate daily returns
daily_returns = np.diff(stock_prices) / stock_prices[:-1]
print("Daily Returns:", daily_returns)
```

```python
# Compute the mean of daily returns
mean_daily_return = np.mean(daily_returns)

# Compute the standard deviation of daily returns
std_daily_return = np.std(daily_returns)

# Compute the variance of daily returns
var_daily_return = np.var(daily_returns)

print(f"Mean Daily Return: {mean_daily_return:.4f}")
print(f"Standard Deviation of Daily Returns: {std_daily_return:.4f}")
print(f"Variance of Daily Returns: {var_daily_return:.4f}")
```

```python
# Identify the day with the highest return
max_return_index = np.argmax(daily_returns)
max_return_value = daily_returns[max_return_index]

# Identify the day with the lowest return
min_return_index = np.argmin(daily_returns)
min_return_value = daily_returns[min_return_index]

print(f"Day with the highest return: Day {max_return_index + 1} ({max_return_value:.4f})")
print(f"Day with the lowest return: Day {min_return_index + 1} ({min_return_value:.4f})")
```

```python
# Calculate the cumulative return over the year
cumulative_return = np.prod(1 + daily_returns) - 1
print(f"Cumulative Return over the year: {cumulative_return:.4f}")
```

### Questin 03

You are working on a project that involves image processing. In this assignment, you will utilize NumPy to implement various operations on an RGB image of your own choice. Follow the steps below:

- Select an Image:
- Choose an RGB image of your own choice. You can use any image. Ensure that the image is in RGB format.
- Convert RGB Image to Grayscale:
- Convert the grayscale image to a binary image using a threshold value.
- Apply Gaussian blur to the binary image to reduce noise.
- Detect edges in the blurred image using the Sobel operator.
- Compute the histogram of pixel intensities in the original grayscale image.


<br> Visualize the results of each operation to observe the effects on the image.
Submission Guidelines:

<br>Submit the selected RGB image in the submission as well.

```python
import cv2

import matplotlib.pyplot as plt

# Load the RGB image
image = cv2.imread('duck.jpg')
# Convert the image from BGR (OpenCV format) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert RGB image to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Display the original RGB image and the grayscale image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
# Convert grayscale image to binary image using a threshold value
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

# Display the binary image
plt.figure(figsize=(5, 5))
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur to the binary image to reduce noise
blurred_image = cv2.GaussianBlur(binary_image, (5, 5), 0)

# Display the blurred image
plt.figure(figsize=(5, 5))
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```

```python
# Detect edges using the Sobel operator
sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)
edges = np.hypot(sobelx, sobely)

# Display the edges
plt.figure(figsize=(5, 5))
plt.imshow(edges, cmap='gray')
plt.title('Edges Detected by Sobel Operator')
plt.axis('off')
plt.show()
```

```python
# Compute the histogram of pixel intensities in the original grayscale image
histogram, bin_edges = np.histogram(gray_image, bins=256, range=(0, 255))

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.plot(bin_edges[0:-1], histogram)
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
