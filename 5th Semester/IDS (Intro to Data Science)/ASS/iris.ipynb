## Data Visualization

```python
import matplotlib.pyplot as plt
```

### 1. Plot a histogram of the 'sepal_length' for the entire dataset.

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=iris, x='sepal_length', bins=20)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

### 2. Create a scatter plot to visualize the relationship between 'sepal_length' and 'sepal_width'.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width (Colored by Species)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```

### 3. Use a boxplot to show the distribution of 'petal_length' for each species.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

### 4. Create a pairplot for all numerical columns, colored by species.

```python
sns.pairplot(data=iris, hue='species')
plt.suptitle('Pairplot of Iris Dataset (Colored by Species)')
plt.show()
```
